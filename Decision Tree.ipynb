```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('C:/Users/yug59/Downloads/tennis.csv')
print(data)
```

```python
X = data.iloc[:,:-1]
print("\nThe first 5 values of Train data is \n",X.head())
```

```python
y = data.iloc[:,-1]
print("\nThe first 5 values of Train output is \n",y.head())
```

```python
le_outlook = LabelEncoder()
X['outlook'] =  le_outlook.fit_transform(X.outlook)
le_Temperature = LabelEncoder()
X['temp'] =  le_Temperature.fit_transform(X.temp)
le_Humidity = LabelEncoder()
X['humidity'] =  le_Humidity.fit_transform(X.humidity)
le_Windy = LabelEncoder()
X['windy'] =  le_Windy.fit_transform(X.windy)
print("\nNow the Train data is\n",X.head())
```

```python
le_PlayTennis = LabelEncoder()
y =  le_PlayTennis.fit_transform(y)
print("\nNow the Train data is\n",y)
```

```python
from sklearn.model_selection import cross_val_predict,KFold,cross_val_score,train_test_split, learning_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
clf = DecisionTreeClassifier(criterion='entropy', random_state=0)
clf=clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
tree.plot_tree(clf)
```
